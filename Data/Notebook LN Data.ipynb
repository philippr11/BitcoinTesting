```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("final_ln_transactions.csv")
df.head()
```

```python
print(f"Durchschnittliche Gebühr: {df['Fee (sats)'].mean():.2f} Sats")
print(f"Minimale Gebühr: {df['Fee (sats)'].min()} Sats")
print(f"Maximale Gebühr: {df['Fee (sats)'].max()} Sats")
```

```python
plt.figure(figsize=(10,5))
plt.hist(df["LN Duration (s)"], bins=30, edgecolor="black", color = "orange")
plt.xlabel("Dauer (s)")
plt.ylabel("Anzahl der Transaktionen")
plt.title("Zeitliche Verteilung der LN-Transaktionsdauer")

plt.savefig("ln_durchschnittsdauer_transaktionen.pdf", bbox_inches="tight")
plt.savefig("ln_durchschnittsdauer_transaktionen.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x="Num Hops", y="Fee (sats)")
plt.xlabel("Anzahl der Hops")
plt.ylabel("Gebühr (sats)")
plt.title("Zusammenhang zwischen Anzahl der Hops und Gebühr")
plt.show()
```

Ein Zusammenhang zwischen mehr Hops = höhere Gebühr ist hier leider nicht ablesbar, obwohl dies zu erwarten wäre
